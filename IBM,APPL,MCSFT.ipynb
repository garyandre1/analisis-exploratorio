```python
import pandas as pd
import yfinance as yf
import matplotlib.pyplot as plt
```

```python
yf.Ticker('AAPL').info
```

```python
tickers = ("AAPL", "MSFT","IBM")
```

```python
infos = []

for i in tickers:
  infos.append(yf.Ticker(i).info)
```

```python
Fundamentals = ["dividendYield","marketCap","beta","forwardPE"]
```

```python
df = pd.DataFrame(infos)
```

```python
df = df.set_index("symbol")
```

```python
df
```

```python
df[df.columns[df.columns.isin(Fundamentals)]]
```

```python
plt.bar(df.index,df.dividendYield,color=("k", "orange","cyan"))
plt.show
```

```python
plt.bar(df.index,df.forwardPE,color=("k", "orange","cyan"))
plt.show
```
