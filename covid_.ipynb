```python
df = pd.read_csv("https://raw.githubusercontent.com/owid/covid-19-data/master/public/data/owid-covid-data.csv")
```

```python
df.head()
```

```python
df.columns
```

```python
date_string = df["date"]

DF_Date = df["date"].fillna(0)
DF_Cases = df["new_cases"].fillna(0)
DF_deaths = df["new_deaths"].fillna(0)
DF_Continents = df["continent"].fillna(0)
DF_country = df["location"].fillna(0)
DF_people_vaccinated = df["new_people_vaccinated_smoothed"].fillna(0)

DF_total = pd.concat([DF_Date, DF_Continents, DF_country,DF_Cases,DF_deaths, DF_people_vaccinated], axis=1)

DF_total.columns = ["Fecha", "Continente", "Pais", "Total contagiados", "Total_fallecidos","Total vacunados"]
```

```python
DF_total.head()
```

```python
DF_total = DF_total[DF_total["Pais"] != "World"]
DF_total = DF_total[DF_total["Pais"] != "Internacional"]
DF_total = DF_total[DF_total["Pais"] != "Europe"]
DF_total = DF_total[DF_total["Pais"] != "North America"]
DF_total = DF_total[DF_total["Pais"] != "Asia"]
DF_total = DF_total[DF_total["Pais"] != "South America"]
DF_total = DF_total[DF_total["Pais"] != "Africa"]
DF_total = DF_total[DF_total["Pais"] != "European Union"]
DF_total = DF_total[DF_total["Pais"] != "South Africa"]
DF_total = DF_total[DF_total["Pais"] != "Oceania"]
DF_total = DF_total[DF_total["Pais"] != "High income"]
DF_total = DF_total[DF_total["Pais"] != "Upper middle income"]
DF_total = DF_total[DF_total["Pais"] != "Lower middle income"]
DF_total = DF_total[DF_total["Pais"] != "Low income"]

DF_total.head()
```

```python
Total_Vacunados = DF_total["Total vacunados"].sum()
Total_Vacunados
```

```python
chileVac = DF_total[DF_total["Pais"] == "Chile"].sum()
chileVac
```

```python
DF_total_melt = pd.melt(DF_total, id_vars=["Fecha", "Continente","Pais"], value_vars=["Total contagiados", "Total vacunados"], var_name = "Estado", value_name="Total")
DF_total_melt.info()
```
