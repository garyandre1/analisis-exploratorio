Analisis de el cierre de acciones para apple y tesla (2010-2020)


```python
import pandas as pd 
import matplotlib.pyplot as plt 
%matplotlib inline 
import yfinance as yf 
import numpy as np 
import seaborn as sns
```

```python
stock = ["AAPL","XPEV"]

stocks = yf.download(stock, start= "2010-01-01", end = "2020-11-01" )
data = stocks.loc[:, "Close"].copy()
```

```python
stock
```

```python
data.plot(figsize = (17,8), fontsize = 18)
plt.style.use("seaborn")
plt.show()
```

Hacemos una vista general de XPEV de todos los sdatos que tenemos .


```python
sns.pairplot(stock_Xpev, height=2.5)
plt.show()
```

```python
index_dates = ['2020-08-27', '2020-08-28', '2020-08-31', '2020-09-01', '2020-09-02']
stock_Xpev = pd.DataFrame(stock_Xpev, index=pd.to_datetime(index_dates))

stock_Xpev_reset = stock_Xpev.reset_index()

```

```python
f, ax = plt.subplots(figsize=(9,9))
sns.barplot(x='index', y='Open', data=stock_Xpev_reset, color='blue', label='Open')
sns.barplot(x='index', y='Close', data=stock_Xpev_reset, color='orange', label='Close')

plt.legend()
plt.show()
```
